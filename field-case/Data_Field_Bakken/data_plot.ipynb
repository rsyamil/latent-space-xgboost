```python
#plotting functions
def plot_wells(data, controls, fileno, idxs):
    f = plt.figure(figsize=[14, 14])
    for i, idx in enumerate(idxs):
        ax1 = plt.subplot(5, 3, i+1)

        _ = data[idx]
        ax1.plot(_[:, 2], c='r', label='Gas', alpha=0.8)
        ax1.plot(_[:, 1], c='b', label='Water', alpha=0.8)
        ax1.plot(_[:, 0], c='g', label='Oil', alpha=0.8)
        ax1.legend()
        ax1.set_xlabel('Timesteps')
        ax1.set_ylabel('Rate')
        ax1.set_title(fileno[idx])

        ax2 = ax1.twinx() 
        c = 'tab:purple'

        #plot control on another axis
        _ = controls[idx]
        ax2.plot(_, color=c)
        ax2.set_ylabel('Control', color=c) 
        ax2.tick_params(axis ='y', labelcolor=c)
        
        ax1.grid(False)
        ax2.grid(False)

    plt.tight_layout()
    f.savefig('readme/sample_profiles.png', dpi=300, bbox_inches='tight')

#plot cumulatives histogram
def plot_hists(data):
    f = plt.figure(figsize=[10, 3.5])
    
    clrs = ['green', 'blue', 'red']
    lbls = ['Oil', 'Water', 'Gas']
    titles = ['Cumm Oil', 'Cumm Water', 'Cumm Gas']
    
    for i in range(len(lbls)):
        plt.subplot(1, 3, i+1)
        bb = np.linspace(np.min(data[:, i]), np.max(data[:, i]), 100)
        plt.hist(data[:, i], color=clrs[i], alpha=0.4, bins=bb, label=lbls[i])
        plt.title(titles[i])
        plt.ylabel("Freq.")
        plt.legend()
        plt.grid(False)
    plt.tight_layout()
    f.savefig('readme/cumm_hists.png', dpi=300, bbox_inches='tight')

#plot properties histogram 
def plot_prop_hists(data):
    f = plt.figure(figsize=[10, 6])
    
    titles = ['Vol', 'Lbs_Prop', 'MTPress', 'MTRate', 'Stages']
    
    for i in range(data.shape[1]):
        plt.subplot(2, 3, i+1)
        bb = np.linspace(np.min(data[:, i]), np.max(data[:, i]), 100)
        plt.hist(data[:, i], color='b', alpha=0.4, bins=bb)
        plt.title(titles[i])
        plt.ylabel("Freq.")
        plt.grid(False)
    plt.tight_layout()
    f.savefig('readme/props_hists.png', dpi=300, bbox_inches='tight')
    
#plot cumulatives bubbles
def plot_bubbles(loc, cumm):
    f = plt.figure(figsize=[10, 30])
    
    clrs = ['green', 'blue', 'red']
    lbls = ['Oil', 'Water', 'Gas']
    titles = ['Cumm Oil', 'Cumm Water', 'Cumm Gas']
    
    for i in range(len(lbls)):
        plt.subplot(3, 1, i+1)
        
        plt.scatter(loc[:,0], loc[:,1], s=cumm[:, i]*0.001, c=clrs[i], alpha=0.3, label=lbls[i])
        plt.scatter(loc[:,0], loc[:,1], s=3, c=clrs[i], alpha=0.8)
        
        plt.xlabel("Latitude")
        plt.ylabel("Longitude")
        plt.axis('scaled')
        plt.title(titles[i])
        plt.legend()
        plt.grid(False)
    plt.tight_layout()
    f.savefig('readme/bubble_maps.png', dpi=300, bbox_inches='tight')
```

```python
#plot cumulatives GOR WCUT bubbles
def plot_bubbles_ratio(loc, cumm):
    f = plt.figure(figsize=[10, 30])
    
    clrs = ['blue', 'green', 'red']
    lbls = ['WCUT', 'GOR', 'GWR']
    titles = ['WCUT', 'GOR', 'GWR']
    
    for i in range(len(lbls)):
        plt.subplot(3, 1, i+1)
        
        plt.scatter(loc[:,0], loc[:,1], s=cumm[:, i]*1000, c=clrs[i], alpha=0.3, label=lbls[i])
        plt.scatter(loc[:,0], loc[:,1], s=3, c=clrs[i], alpha=0.8)
        
        plt.xlabel("Latitude")
        plt.ylabel("Longitude")
        plt.axis('scaled')
        plt.title(titles[i])
        plt.legend()
        plt.grid(False)
    plt.tight_layout()
    f.savefig('readme/bubble_maps_ratio.png', dpi=300, bbox_inches='tight')
    
#function to normalize cumms by features: [num_data, features]
def normalizeCumm(data):
    output_maxs = np.zeros([data.shape[1],])
    output_maxs[0] = np.max(data[:, 0])
    output_maxs[1] = np.max(data[:, 1]) 
    output_maxs[2] = np.max(data[:, 2]) 
    return (data/(output_maxs)), output_maxs
```

```python
#calculate WCUT and GOR and GWR

#normalize cumulative data by channel
Cumm_norm_, Cumm_maxs = normalizeCumm(Cumm_)

Cumm_ratio_ = np.copy(Cumm_norm_.astype("float"))

Cumm_ratio_[:,0] = Cumm_norm_[:,1]/(Cumm_norm_[:,0]+Cumm_norm_[:,1])
Cumm_ratio_[:,1] = Cumm_norm_[:,2]/(Cumm_norm_[:,0]+Cumm_norm_[:,2])
Cumm_ratio_[:,2] = Cumm_norm_[:,2]/(Cumm_norm_[:,1]+Cumm_norm_[:,2])

plot_bubbles_ratio(Loc_, Cumm_ratio_)
```

```python
#2d interpolation to see if trend exists
from scipy import interpolate

#plot cumulatives GOR WCUT contours
def contour_bubbles_ratio(loc, cumm):
    f = plt.figure(figsize=[10, 30])
    
    clrs = ['blue', 'green', 'red']
    lbls = ['WCUT', 'GOR', 'GWR']
    titles = ['WCUT', 'GOR', 'GWR']
    
    for i in range(len(lbls)):
        plt.subplot(3, 1, i+1)
        
        print(np.min(cumm[:, i]), np.max(cumm[:, i])) 
 
        f0 = interpolate.interp2d(loc[:,0], loc[:,1], cumm[:, i], kind='cubic')
        xnew = np.arange(np.min(loc[:,0]), np.max(loc[:,0]), 1e-3)
        ynew = np.arange(np.min(loc[:,1]), np.max(loc[:,1]), 1e-3)
        znew = f0(xnew, ynew)
        
        #truncate large, small values
        print(np.min(znew), np.max(znew))
        znew[znew <= 0] = 0
        znew[znew >= 1] = 1
        print(np.min(znew), np.max(znew)) 
        
        CS = plt.contour(xnew, ynew, znew, levels=30, cmap='jet')
        plt.clabel(CS, inline=True, fontsize=10)

        #plt.scatter(loc[:,0], loc[:,1], s=3, c=clrs[i], alpha=0.8, label=lbls[i])
        
        plt.xlabel("Latitude")
        plt.ylabel("Longitude")
        plt.axis('scaled')
        plt.title(titles[i])
        plt.legend()
        plt.grid(False)
    plt.tight_layout()
    f.savefig('readme/contour_maps_ratio.png', dpi=300, bbox_inches='tight')
```

```python
#plot fields
def plot_fields(loc, fieldname):
    #get number of unique fields
    unique_fields = np.unique(fieldname)
    no_unique_fields = len(unique_fields)
    field_code = np.arange(no_unique_fields)

    my_cmap = cm.get_cmap('Dark2')
    my_norm = Normalize(vmin=0, vmax=(no_unique_fields-1))
    cs = my_cmap(my_norm(field_code))

    #assign each data point the field code
    FieldName_idx = np.zeros(FieldName_.shape, dtype='int32')
    for i in range(len(FieldName_)):
        idx, = (np.where(unique_fields == FieldName_[i]))[0]
        FieldName_idx[i] = idx

    f = plt.figure(figsize=[10, 10])
    plt.scatter(loc[:,0], loc[:,1], s=50, c=cs[FieldName_idx])

    plt.xlabel("Latitude")
    plt.ylabel("Longitude")
    plt.axis('scaled')
    plt.title(str(no_unique_fields) + " unique fields")
    plt.grid(False)
    plt.tight_layout()
    f.savefig('readme/field_maps.png', dpi=300, bbox_inches='tight')
```

```python
plot_fields(Loc_, FieldName_)
```
